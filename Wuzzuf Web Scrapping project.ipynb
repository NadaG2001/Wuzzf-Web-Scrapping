```python
import requests
import csv
from bs4 import BeautifulSoup
from itertools import zip_longest
```

```python
jop_title =[]
company_name =[]
location =[]
jop_skill =[]
links =[]
#jop_time =[]
```

### use requests to fetch the url

```python
result = requests.get("https://wuzzuf.net/search/jobs/?q=python+Developer&a=hpb")
```

### save Page Content

```python
src = result.content
src
```

### create soup object to parse content

```python
soup = BeautifulSoup(src,"lxml")
soup
```

### find Elements containing info we need

-- jop titles, jop skills, company names, locations

```python
jop_titles = soup.find_all("h2", {"class":"css-m604qf"})
jop_titles
```

```python
company_names =soup.find_all("a",{"class":"css-17s97q8"})
company_names
```

```python
locations = soup.find_all("span",{"class","css-5wys0k"})
locations
```

```python
jop_skills = soup.find_all("div",{"class","css-y4udm8"})
jop_skills
```

### looping to return lists to extract needed info into other lists

```python
for i in range(len(jop_titles)):
    jop_title.append(jop_titles[i].text)
    links.append(jop_titles[i].find("a").attrs['href'])
    company_name.append(company_names[i].text)
    location.append(locations[i].text)
    jop_skill.append(jop_skills[i].text)
    #jop_time.append(jop_Time[i].text)
print(jop_title,company_name,location,jop_skill, links)
```

### create csv file and fill it with values

```python
file_list = [jop_title, company_name, location, jop_skill ,links]
exported = zip_longest(*file_list)    # * for unbaking
with open("Wuzzf_Data.csv","w") as myfile:
    wr = csv.writer(myfile)
    wr.writerow(["Job title", "Company name","Location","Skills","Links"])
    wr.writerows(exported)
```
